## Preprocessing

```python
xn_train = x_train.drop(['date', 'train', 'way', 'station', 'hour', 'composition'], axis = 1)
yn_train = y_train.drop(['Unnamed: 0'], axis = 1)
xn_test = x_test.drop(['date', 'train', 'way', 'station', 'hour', 'composition'], axis = 1)
```

```python
model = HistGradientBoostingRegressor(loss = 'absolute_error', learning_rate = 0.1, max_leaf_nodes = 7)
```

```python
model.fit(xn_train,yn_train)
y_pred = model.predict(xn_test)
```

```python
y_predict = pd.DataFrame({"p0q0" : y_pred})
```
